```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.random.seed(42)

T = 1500
time = np.arange(T)

trend = 0.01 * time
seasonality = 10 * np.sin(2 * np.pi * time / 50)

# Heteroscedastic noise (variance increases)
noise = np.random.normal(0, 1 + 0.005*time)

feature1 = trend + seasonality + noise

# Regime change
feature1[800:] += 20

feature2 = np.roll(feature1, 3) + np.random.normal(0, 1, T)
feature3 = np.sqrt(np.abs(feature1)) + np.random.normal(0, 0.5, T)

target = (
    0.5 * feature1 +
    0.3 * feature2 -
    0.2 * feature3 +
    np.random.normal(0, 1, T)
)

data = pd.DataFrame({
    "f1": feature1,
    "f2": feature2,
    "f3": feature3,
    "y": target
})

data.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(data["y"])
plt.title("Target Variable")
plt.show()
```

```python
def create_sequences(data, seq_len=30):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data.iloc[i:i+seq_len][["f1","f2","f3"]].values)
        y.append(data.iloc[i+seq_len]["y"])
    return np.array(X), np.array(y)

seq_len = 30
X, y = create_sequences(data, seq_len)
```

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_reshaped = X.reshape(-1, 3)
X_scaled = scaler_X.fit_transform(X_reshaped)
X_scaled = X_scaled.reshape(X.shape)

y_scaled = scaler_y.fit_transform(y.reshape(-1,1))
```

```python
train_size = 1000

X_train = X_scaled[:train_size]
y_train = y_scaled[:train_size]

X_test = X_scaled[train_size:]
y_test = y_scaled[train_size:]
```

```python
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)

X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)

# Baseline Model - Linear Regression

from sklearn.linear_model import LinearRegression

# Reshape for sklearn
X_train_lr = X_train.reshape(X_train.shape[0], -1)
X_test_lr = X_test.reshape(X_test.shape[0], -1)

lr = LinearRegression()
lr.fit(X_train_lr, y_train)

pred_lr = lr.predict(X_test_lr)

# Convert back to original scale
pred_lr = scaler_y.inverse_transform(pred_lr)
y_test_actual = scaler_y.inverse_transform(y_test.numpy())

mae_lr = mean_absolute_error(y_test_actual, pred_lr)
rmse_lr = np.sqrt(mean_squared_error(y_test_actual, pred_lr))

print("Baseline MAE:", mae_lr)
print("Baseline RMSE:", rmse_lr)
```

```python
class AttentionModel(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.attn = nn.Linear(hidden_dim, 1)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attn(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        output = self.fc(context)
        return output, attn_weights

model = AttentionModel()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 30

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output, _ = model(X_train)
    loss = criterion(output, y_train)
    loss.backward()
    optimizer.step()

    if (epoch+1) % 5 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```

```python
model.eval()
with torch.no_grad():
    preds, attn_weights = model(X_test)

preds = scaler_y.inverse_transform(preds.numpy())
y_test_actual = scaler_y.inverse_transform(y_test.numpy())

mae = mean_absolute_error(y_test_actual, preds)
rmse = np.sqrt(mean_squared_error(y_test_actual, preds))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(12,4))
plt.plot(y_test_actual[:200], label="Actual")
plt.plot(preds[:200], label="Predicted")
plt.legend()
plt.show()
```

```python
avg_attention = attn_weights.mean(dim=0).squeeze().numpy()

plt.plot(avg_attention)
plt.title("Average Attention Weights")
plt.show()
```
